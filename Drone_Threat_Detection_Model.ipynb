```python
from ultralytics import YOLO


model = YOLO('yolov8n.pt')


results = model.train(
    data='/content/drive/MyDrive/DroneThreatProject/dataset/data.yaml',  # path
    epochs=50,         # training length
    imgsz=640,         # image resolution
    batch=8,           # adjust memory errors
    name='drone_model_v1',  # model folder name under /runs/detect/
    workers=4,         # parallel data loading
    device=0,          # use GPU if available
    exist_ok=True,     # overwrite previous runs
    verbose=True       # show training logs
)
```
